```python
# Import libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Create a simple dataset
data = {
    'Age': [25, 45, 30, 35, 40, 22, 50, 23, 38, 48],
    'Annual_Income': [30000, 60000, 35000, 45000, 50000, 25000, 70000, 28000, 52000, 68000],
    'Spending_Score': [60, 40, 65, 50, 45, 80, 30, 75, 48, 35]
}

# Convert to DataFrame
df = pd.DataFrame(data)
print("Customer Data:\n", df)

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['Age', 'Annual_Income', 'Spending_Score']])

# Display cluster results
print("\nCustomer Segments:\n", df)

# Visualize clusters
plt.figure(figsize=(8,6))
plt.scatter(df['Annual_Income'], df['Spending_Score'], c=df['Cluster'], cmap='rainbow', s=100)
plt.title('Customer Segmentation using K-Means')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
