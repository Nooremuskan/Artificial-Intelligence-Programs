```python
#Data Preprocessing Step's

import pandas as pd
import numpy as np

# Create the DataFrame
data = {
    'Name': ['Asha', 'Ravi', 'Kiran', 'Lata', 'Meena', 'Raj'],
    'Age': [20, 21, np.nan, 23, 22, np.nan],
    'Marks': [85, 90, np.nan, 70, 88, 92],
    'Gender': ['F', 'M', 'M', np.nan, 'F', 'M']
}

df = pd.DataFrame(data)
print("Original Data:\n", df)

# Clean data: convert names to lowercase
df['Name'] = df['Name'].str.lower()

# Create a new 'Performance' column based on marks
def performance_category(marks):
    if pd.isna(marks):
        return 'No Data'
    elif marks >= 90:
        return 'Excellent'
    elif marks >= 80:
        return 'Good'
    elif marks >= 70:
        return 'Average'
    else:
        return 'Poor'

df['Performance'] = df['Marks'].apply(performance_category)

# Display cleaned DataFrame
print("\nAfter Cleaning and Adding Performance:\n", df)

# Select final columns
df_final = df[['Age', 'Marks', 'Performance']]
print("\nFinal Selected Data:\n", df_final)
```
